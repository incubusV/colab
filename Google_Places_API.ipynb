```
import pandas as pd, numpy as np
import requests
import json
import time

!ls '/content/drive/MyDrive/Colab Notebooks/Data/google_api_key/'

path_file = '/content/drive/MyDrive/Colab Notebooks/Data/google_api_key/'
api_key = open(path_file+'place_api_key.txt', "r")

final_data = []
#coordinates = ['25.063205,121.499688'] # Taiwan
coordinates = ['13.805928,100.523952']  # Gateway Bangsue
keywords = ['restaurant','hotel']
radius = '100'

#insert your API key here
api_key = api_key.readline()

for coordinate in coordinates:
  for keyword in keywords:
    url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+coordinate+'&radius='+str(radius)+'&keyword='+str(keyword)+'&key='+str(api_key)
    print(url)
    i = 1
    while True:
      print(i)
      jj = []
      response = requests.get(url)
      jj = json.loads(response.text)
      results = jj['results']
      #print(results)
      
      for result in results:
        name = result['name']
        place_id = result ['place_id']
        lat = result['geometry']['location']['lat']
        lng = result['geometry']['location']['lng']
        rating = result['rating']
        user_rating = result['user_ratings_total']
        types = result['types']
        unq_type = result['types'][0]
        vicinity = result['vicinity']
        data = [name, place_id, lat, lng, rating, user_rating, types, unq_type, vicinity]
        print(data)
        final_data.append(data)
        time.sleep(2)

      if 'next_page_token' not in jj:
        break
      else:
        next_page_token = jj['next_page_token']
        url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?key='+str(api_key)+'&pagetoken='+str(next_page_token)
        print (url)
        i = i+1
        time.sleep(2)


labels = ['Place_Name','Place_ID', 'Latitude', 'Longitude', 'Rating', 'User Ratings Total', 'Types', 'Unq_Type', 'Vicinity']
export_dataframe_google = pd.DataFrame.from_records(final_data, columns=labels)
export_dataframe_google.to_csv('/content/drive/MyDrive/Colab Notebooks/Data/export_dataframe_google.csv')

print('#####################  FINISH  ########################')

      
```

```
import pandas as pd
import folium
from folium import FeatureGroup, LayerControl
#from folium.plugins import TimestampedGeoJson
import json
#import matplotlib.pyplot as plt

# Read Data
path_file = '/content/drive/MyDrive/Colab Notebooks/Data/'

google_places_data = pd.read_csv(path_file +'export_dataframe_google.csv')
google_places_data.info()

### Create Map ###
google_places_map = folium.Map(location=[13.805928,100.523952], zoom_start=13)

for index, place in google_places_data.iterrows():
    folium.Marker(location=[place.Latitude, place.Longitude], popup=place.Place_Name).add_to(google_places_map)

# google_places_map
```

```
places_types = list(google_places_data.Unq_Type.unique())
#print(places_types)

#colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4', '#46f0f0', '#f032e6', '#bcf60c', '#cb93d2', '#008080', '#e6beff', '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1', '#000075', '#808080', '#ffffff', '#000000']

colors = ['#3cb44b', '#e6194b', '#ffe119', '#4363d8', '#f58231']


google_places_map = folium.Map(location=[13.805928,100.523952], zoom_start=15)

ptypes_groups = {}

for ptype in places_types:
  ptypes_groups[ptype] = FeatureGroup(name=ptype)


for index, place in google_places_data.iterrows():
  folium.Circle(location=[place.Latitude, place.Longitude], popup=place.Place_Name + ' ' + place.Unq_Type, stroke=False
              ,fill=True,radius=int(place.Rating*8), fill_color=colors[places_types.index(place.Unq_Type)]
              ,fill_opacity=0.9).add_to(ptypes_groups[place.Unq_Type])

              

for ptype, group in ptypes_groups.items():
  #print(ptype)
  group.add_to(google_places_map)

LayerControl().add_to(google_places_map)

google_places_map
```
